```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
%pylab inline
```

```python
Samsung = pd.read_csv("005930.KS.csv")
Apple = pd.read_csv("AAPL.csv")
Amazon = pd.read_csv("AMZN.csv")
Alibaba = pd.read_csv("BABA.csv")
Facebook = pd.read_csv("FB.csv")
Google = pd.read_csv("GOOG.csv")
Microsoft = pd.read_csv("MSFT.csv")
Nvidia = pd.read_csv("NVDA.csv")
Tencent = pd.read_csv("TCEHY.csv")
TSMC = pd.read_csv("TSMC34.SA.csv")

stock = [Samsung, Apple, Amazon, Alibaba, Facebook, Google, Microsoft, Nvidia, Tencent, TSMC]
```

```python
for i in stock:
    plt.plot(i.Date, i.Close)
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend(['Samsung', 'Apple', 'Amazon', 'Alibaba', 'Facebook', 'Google', 'Microsoft', 'Nvidia', 'Tencent', 'TSMC'])
plt.show()
```
